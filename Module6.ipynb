```python
#_____6.2_____
mean_daily_returns= df_data.pct_change().mean()
annual_mean=mean_daily_returns * 252
print("Average Annual Return:\n")
print(annual_mean)
```

```python
std_daily_returns=df_data.pct_change().std()
volatility=std_daily_returns * np.sqrt(252)
print("Volatility of returns:\n")
print(volatility)
```

```python
#______6.3______
returns = df_data.pct_change().mean() * 252
returns = pd.DataFrame(returns)
returns.columns = ['Returns']
returns['Volatility'] = df_data.pct_change().std() * np.sqrt(252)


data = np.asarray([np.asarray(returns['Returns']),np.asarray(returns['Volatility'])]).T

X = data
distorsions = []
for k in range(1, 11):
    k_means = KMeans(n_clusters=k)
    k_means.fit(X)
    distorsions.append(k_means.inertia_)

fig = plt.figure(figsize=(15, 5))
plt.plot(range(1, 11), distorsions, marker="o")
plt.grid()
plt.title('Elbow curve')
```

```python
plt.figure(figsize=(10,6))
plt.grid()
data = np.asarray([np.asarray(returns['Returns']),np.asarray(returns['Volatility'])]).T

# computing K-Means with K = 5 (5 clusters)
centroids,_ = kmeans(data,5)

idx,_ = vq(data,centroids)

plot(data[idx==0,0],data[idx==0,1],'ob',label='cluster A')
plot(data[idx==1,0],data[idx==1,1],'om',label='cluster B')
plot(data[idx==2,0],data[idx==2,1],'oy',label='cluster C')
plot(data[idx==3,0],data[idx==3,1],'or',label='cluster D')
plot(data[idx==4,0],data[idx==4,1],'og',label='cluster E')
plt.title("Number Of Clusters")

plt.legend()
plt.tight_layout()
show()
```

```python
#______6.4_______
cluster_details = [(cluster,name) for cluster, name in zip(idx+1 , returns.index)]
cluster_details=pd.DataFrame(cluster_details)
cluster_details.columns=['Cluster' ,'Stock' ]
    
grp = cluster_details.groupby('Cluster') 
for name, group in grp: 
    print("Cluster",name ,":")
    print("\n")
    print(group) 
    print("\n\n") 
```
