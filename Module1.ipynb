```python
df['Month']=df['Date'].dt.month
df['Year']=df['Date'].dt.year
```

```python
df1=df[['Year','Month','Total Traded Quantity']].groupby(by=['Month','Year']).sum()
df2=df[['Year','Month','Turnover']].groupby(by=['Month','Year']).sum()
vwap=df2['Turnover']/df1['Total Traded Quantity']
print(vwap)
```

```python
def avg_price_N_day(N):
    return df['Average Price'][len(df) - N:].mean()

print('The average price over last 1 week is:' , (avg_price_N_day(7)))
print('The average price over last 2 week is:' , (avg_price_N_day(14)))
print('The average price over last 1 month is:' , (avg_price_N_day(30)))
print('The average price over last 3 month is:' , (avg_price_N_day(3*30)))
print('The average price over last 6 month is:' , (avg_price_N_day(6*30)))
print('The average price over last 1 year is:' , (avg_price_N_day(12*30)))
print('\n')

def profit_loss(N):
    return (df['Close Price'][len(df)-N]-df['Close Price'][N])/df['Close Price'][N]*100

print('The Profit/Loss percent over last 1 week is:', profit_loss(7))
print('The Profit/Loss percent over last 2 week is:', profit_loss(14))
print('The Profit/Loss percent over last 1 month is:', profit_loss(30))
print('The Profit/Loss percent over last 3 month is:', profit_loss(30*3))
print('The Profit/Loss percent over last 6 month is:', profit_loss(30*6))
print('The Profit/Loss percent over last 1 year is:', profit_loss(30*12))
```

```python
df['Day_Perc_Change']=df['Close Price'].pct_change()*100
#df['Day_Perc_Change'][0]= 0
df.__setitem__(0, 0.00)
```

```python
def cases(num):
    if num >= -0.5 and num <= 0.5 : return 'Slight or No Change' 
    elif num >= 0.5 and num <=1 : return 'Slight Positive'
    elif num > -1 and num < -0.5: return 'Slight Negative'
    elif num > 1 and num < 3: return 'Positive'
    elif num > -3 and num < -1: return 'Negative'
    elif num > 3 and num < 7: return 'Among Top Gainers'
    elif num > -7 and num < -3: return 'Among Top Losers'
    elif num > 7 : return 'Bull Run'
    elif num  < -7: return 'Bear Drop'
  
df['Trend']= df.Day_Perc_Change.apply(cases)
```
